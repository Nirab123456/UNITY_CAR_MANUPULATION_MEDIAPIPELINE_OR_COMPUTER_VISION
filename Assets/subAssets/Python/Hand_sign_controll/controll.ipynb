```python
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    # # Retrieve the class labels
    # class_labels = results.xyxyn[0][:, -1].cpu().numpy()
    # print(class_labels)
    # print(class_labels.shape)
    
    # # Print detected class names
    # print(class_labels)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
for i in range(100):
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Make detections 
        results = model(frame)
        
        cv2.imshow('YOLO', np.squeeze(results.render()))
        unity_comms.GoForward_5000()    
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()    
```
